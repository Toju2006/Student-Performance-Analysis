```python
import pandas as pd
df = pd.read_csv('messy_sales.csv')
print(df)
```

```python
#Check missing
print(df.isnull().sum())
```

```python
#Fill missing numbers
tj = df['Sales'] = df['Sales'].fillna(df['Sales'].mean())
print(tj)
```

```python

toj = df['Cost'] = df['Cost'].fillna(df['Sales'].mean())
print(toj)
```

```python
fos = df['Category'] = df['Category'].fillna('Unknown')
print(fos)
```

```python
df.drop_duplicates()
```

```python
df['Profit'] = df['Sales'] - df['Cost']
print(df.groupby('Category') ['Sales'].sum())
```

```python
df['Sales'].plot(kind='bar')
```

```python
df.plot(x='Month', y='Sales',kind='line')
```
